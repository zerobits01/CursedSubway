```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df = pd.read_csv("subway_dataset.csv")
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=['stationname', 'datetime'], keep='first')
```

```python
df.describe()
```

```python
grouped = df.groupby('stationname')
new_variables = {group: data for group, data in grouped}
```

```python
new_variables
```

```python
print(f"count of values: {len(new_variables)}, and keys are:\n{new_variables.keys()}")
```

```python
# what should i do now?
'''
first i write the code for one station
then i will run a for loop on new_variables that contains all parts of preprocessing and fit train check

what should be the loss function? i think the best loss function for this can be MSE because we have numeric values to predict


what should be the preprocessing step?
1. splitting the date time
2. i think i can use the hour and minute as inputs and the date wont have any effect on this!
3. so a polynomial regression can handle this project
'''
```

```python
input_keys = ["day", "hour", "minute"]
output_keys = ["input", "output", "peopleinsidetrain"]

## later i add a for loop here
# station = list(new_variables.keys())[0]
i = 0
for station in new_variables.keys():
    print(30*"#")
    print(f"{i}. station name is: {station}")
    i = i + 1
    for ok in output_keys:
        X = new_variables[station][input_keys]
        y = new_variables[station][ok]
        polynomial = PolynomialFeatures(degree=2,include_bias=False)
        poly_features = polynomial.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.2, random_state=42)
        model = LinearRegression(fit_intercept=True)
        model.fit(X_train,y_train)
        predictions = model.predict(X_test)
        
        MAE = mean_absolute_error(y_test,predictions)
        MSE = mean_squared_error(y_test,predictions)
        RMSE = np.sqrt(MSE)
        
        print("MAE of %s: %f" % (ok, MAE))
        print("RMSE of %s: %f" % (ok, RMSE))
        # residuals = y_test - predictions
        # sns.scatterplot(x=y_test,y=residuals)
        # plt.axhline(y=0, color='r', linestyle='--')
        # plt.xlabel("Charges")
        # plt.ylabel("Residuals")
        # plt.title("Residual Plot")
        # plt.show()
```

```python
from sklearn.svm import SVR

## later i add a for loop here
# station = list(new_variables.keys())[0]
i = 0
for station in new_variables.keys():
    print(30*"#")
    print(f"{i}. station name is: {station}")
    i = i + 1
    for ok in output_keys:
        X = new_variables[station][input_keys]
        y = new_variables[station][ok]
        polynomial = PolynomialFeatures(degree=2,include_bias=False)
        poly_features = polynomial.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.2, random_state=42)
        model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
        model.fit(X_train,y_train)
        predictions = model.predict(X_test)
        
        MAE = mean_absolute_error(y_test,predictions)
        MSE = mean_squared_error(y_test,predictions)
        RMSE = np.sqrt(MSE)
        
        print("MAE of %s: %f" % (ok, MAE))
        print("RMSE of %s: %f" % (ok, RMSE))
        residuals = y_test - predictions
        # sns.scatterplot(x=y_test,y=residuals)
        # plt.axhline(y=0, color='r', linestyle='--')
        # plt.xlabel("Charges")
        # plt.ylabel("Residuals")
        # plt.title("Residual Plot")
        # plt.show()
```
